```python
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings

loader = PyPDFLoader('/kaggle/input/pdffile/n.-gregory-mankiw-macroeconomics-7th-edition-2009.pdf')
documents = loader.load()
```

```python
import nltk

nltk.download('punkt')
```

```python
def word_count(text):
    words = nltk.word_tokenize(text)
    return len(words)

# Filter documents
filtered_documents = [doc for doc in documents if word_count(doc.page_content) >= 50]
```

```python
md = documents[39:]
```

```python
filtered_documents = [doc for doc in md if word_count(doc.page_content) >= 50]
```

```python
for doc in filtered_documents:
    doc.page_content = doc.page_content.replace('\n', ' ')
```

```python
import pandas as pd
data = []
for doc in filtered_documents:
    page_data = {
        'page_content': doc.page_content,
        'word_count': word_count(doc.page_content),
        # Add any other metadata here if needed
    }
    data.append(page_data)

# Convert to DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings.sentence_transformer import (SentenceTransformerEmbeddings)

embedder = SentenceTransformerEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text_splitter = CharacterTextSplitter(
    separator=" ",
    chunk_size=1000,
    chunk_overlap=10
)
docs = text_splitter.split_documents(filtered_documents)
```

```python
db = Chroma.from_documents(docs, embedder)
```

```python
from langchain.llms import HuggingFaceHub


repo_id = "mistralai/Mixtral-8x7B-Instruct-v0.1"
llm = HuggingFaceHub(
  repo_id=repo_id, 
  model_kwargs={"temperature": 0.8, "top_k": 50}, 
  huggingfacehub_api_token='' # enter your token
)
```
